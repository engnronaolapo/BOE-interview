### ANALYSIS

There are 17 metrics in both dataset combined but for the purpose of this analysis, onle five (5) will be considerd. These include 

      1. Gross Written Premium (GWP) – total revenue written by an insurer. Equivalent of turnover for a non-insurance firm.
      2. Net Written Premium (NWP) – GWP less reinsurance. NWP / GWP will show how much of the firm’s risk is being passed                on to reinsurers.
      3. SCR coverage ratio – a measure of whether a firm is meeting its prudential capital requirements. Greater than 100%              means the firm is holding enough capital to meet the requirement. The size of the buffer (i.e. surplus over 100%) can            be important
      4. Gross claims incurred – a large cost to an insurer. Monitoring how these change over time for a firm is vital.
      5 Net combined ratio – (incurred losses plus expenses) / earned premiums. This is a ratio that can indicate the                  profitability of a firm. If this is less than 100% it indicates a profit.

The data for each of these metrics was recorded for over 5 years. Therefore, analysing onle one year out of the five might not be a true representation of the metric. for this reason, the average of these  metrics will be used in these analysis.

An average is a single number taken as representative of a list of numbers, usually the sum of the numbers divided by how many numbers are in the list.
The Data containing each of the metrics is being pulled out to form another dataframe and anylysis is being done on this new datafram.

###  (II)   FURTHER INSIGHTS INTO THE DATA 

To get further insights into the data, the 5 dataframes used to generate the analysis above will be merged togerther. these include df3, df4, df5, df6 and df8 

```python
# merge the dataframes into one big dataframe
from functools import reduce
dfs = [df3,df4,df5,df6,df7]
df8 = reduce(lambda  left,right: pd.merge(left,right,on=['Firm ID'], how='outer'), dfs)
df8
```

It can be seen from the above output above that the merging is still in line as the resulting dataframe contains 31 coulumns (five metrics over the 5years considered and their respective averages) and 325 rows  
For the sake of simplicity again,  the averages of the metrics will be used 

```python
#  creating a dataframe from the averages of the metrica considered
df9 =df8[['Firm ID', 'Average GWP(£m)','Average NWP(£m)','Average SCR coverage ratio','Average Gross claims incurred (£m)','Average Net combined ratio']].copy()
df9
```

```python
# To determine how these averages correlate with eachother 
corr = df9.corr()
corr
```

Correlation makes it clear how one or more variables are related to each other. These variables in these case are the averages of the metrics considered over the 5 year under consideration 

Correlation is statistical technique which determines how one variables moves/changes in relation with the other variable. It gives us the idea about the degree of the relationship of the two variables. It’s a bi-variate analysis measure which describes the association between different variables
 as it can be seen above the possible range of values for the correlation coefficient is -1.0 to 1.0. In other words, the values cannot exceed 1.0 or be less than -1.0. A correlation of -1.0 indicates a perfect negative correlation, and a correlation of 1.0 indicates a perfect positive correlation. If the correlation coefficient is greater than zero, it is a positive relationship. Conversely, if the value is less than zero, it is a negative relationship. A value of zero indicates that there is no relationship between the two variables.
 
The pattern of correlation can also be shown with a heat map

```python
# to plot the heatmap showing the correlation pattern 
plt.subplots(figsize=(10,10))
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=90,
    horizontalalignment='right'
);
```
